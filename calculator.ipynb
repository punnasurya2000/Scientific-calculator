```python
# basic calculator using python
import tkinter as tk

def button_click(number):
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(0, current + str(number))

def button_clear():
    entry.delete(0, tk.END)

def button_equal():
    try:
        result = eval(entry.get())
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except Exception as e:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

# Create a tkinter window
window = tk.Tk()
window.title("Simple Calculator")

# Create an entry widget for input/output
entry = tk.Entry(window, width=45, borderwidth=10)
entry.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

# Define buttons
buttons = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
    ('0', 4, 0), ('C', 4, 1), ('=', 4, 2), ('+', 4, 3)
]

# Create and place buttons in the grid
for (text, row, col) in buttons:
    if text == '=':
        button = tk.Button(window, text=text, padx=40, pady=20, command=button_equal)
    elif text == 'C':
        button = tk.Button(window, text=text, padx=40, pady=20, command=button_clear)
    else:
        button = tk.Button(window, text=text, padx=40, pady=20, command=lambda num=text: button_click(num))
    button.grid(row=row, column=col)

# Run the tkinter event loop
window.mainloop()
```

```python
# added some more functionality to make it scientific calculator

import tkinter as tk
import math

def button_click(number):
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(0, current + str(number))

def button_clear():
    entry.delete(0, tk.END)

def button_equal():
    try:
        result = eval(entry.get())
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except Exception as e:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

def button_pi():
    entry.insert(tk.END, math.pi)

def button_e():
    entry.insert(tk.END, math.e)

def button_sqrt():
    entry.insert(tk.END, "math.sqrt(")

def button_log():
    entry.insert(tk.END, "math.log10(")

def button_ln():
    entry.insert(tk.END, "math.log(")

def button_sin():
    entry.insert(tk.END, "math.sin(")

def button_cos():
    entry.insert(tk.END, "math.cos(")

def button_tan():
    entry.insert(tk.END, "math.tan(")

# Create a tkinter window
window = tk.Tk()
window.title("Scientific Calculator")

# Create an entry widget for input/output
entry = tk.Entry(window, width=35, borderwidth=5)
entry.grid(row=0, column=0, columnspan=6, padx=10, pady=10)

# Define buttons
buttons = [
    ('Ï€', button_pi), ('e', button_e), ('sqrt', button_sqrt), ('log', button_log),
    ('7', lambda: button_click(7)), ('8', lambda: button_click(8)), ('9', lambda: button_click(9)), ('/', lambda: button_click('/')),
    ('4', lambda: button_click(4)), ('5', lambda: button_click(5)), ('6', lambda: button_click(6)), ('*', lambda: button_click('*')),
    ('1', lambda: button_click(1)), ('2', lambda: button_click(2)), ('3', lambda: button_click(3)), ('-', lambda: button_click('-')),
    ('0', lambda: button_click(0)), ('C', button_clear), ('=', button_equal), ('+', lambda: button_click('+')),
    ('sin', button_sin), ('cos', button_cos), ('tan', button_tan), ('ln', button_ln)
]

# Create and place buttons in the grid
row = 1
col = 0
for (text, command) in buttons:
    button = tk.Button(window, text=text, padx=20, pady=20, command=command)
    button.grid(row=row, column=col, padx=2, pady=2)
    col += 1
    if col > 3:
        col = 0
        row += 1

# Run the tkinter event loop
window.mainloop()
```
